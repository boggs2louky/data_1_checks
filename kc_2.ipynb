```python
import pandas as pd 
import seaborn as sns
```

```python
IC = pd.read_csv("https://raw.githubusercontent.com/boggs2louky/data_1_checks/main/assets/HHLPI.csv")
IC
```

```python
#excluding columns individually
IC = (IC.loc[:, IC.columns != 'hlpi_name'])
IC = (IC.loc[:, IC.columns != 'series_ref'])
IC = (IC.loc[:, IC.columns != 'hlpi'])
IC = (IC.loc[:, IC.columns != 'nzhec'])
IC = (IC.loc[:, IC.columns != 'nzhec_name'])
IC = (IC.loc[:, IC.columns != 'level'])
IC = (IC.loc[:, IC.columns != 'change.q'])
IC = (IC.loc[:, IC.columns != 'change.a'])

IC
```

```python
#selecting item group from the 'nzhec_short' group, selecting "Food"
IC = IC[IC['nzhec_short'].str.contains('Food')]
IC
```

```python
# Scatterplot using seaborn, for Index price of Foods in NZ from Quarter 2 of 2008 - Quarter 3 of 2022.
#time_IC = IC.groupby('quarter') ['index'].sum()
#time_IC
```

```python
sns.set(rc = {'figure.figsize' : (15, 10)})
ax = sns.lineplot(x = 'quarter', y = 'index', data = IC)
ax.tick_params(axis = 'x', rotation = 90)
#sns.scatterplot(x = IC['quarter'], y = IC['index']) didn't want to do a scatterplot
```
